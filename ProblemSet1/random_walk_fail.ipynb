```python
#The confidence interval contains 0, so we cannot reject the null hypothesis that the mean is 0.
import numpy as np
import scipy.stats as stats

# Given data
mean_xend = -1.072
std_xend = 31.585610901168273
n = len(xend)  # Number of samples

# Standard error of the mean (SEM)
sem = std_xend / np.sqrt(n)

# Desired confidence level
confidence_level = 0.95

# Critical value from the normal distribution (z-score)
z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)
print(z_critical)

# Margin of error
margin_of_error = z_critical * sem

# Confidence interval
ci_lower = mean_xend - margin_of_error
ci_upper = mean_xend + margin_of_error

print(f"95% Confidence interval: ({ci_lower}, {ci_upper})")
```

We get the normal distribution, because each trajectory is a sum of 1001 random variables with values 1 and -1. Each trajectory is a sample.

Each particle follows a different trajectory, because it flips a coin independently for each particle at each timestep. 
In other words, whether the particle moves in a negative or positive direction at each time point is random.

```python
def mean_displacement_t(X):
    #broadcasting
    return np.mean(np.abs(X), axis=0)
print(mean_displacement_t(dat['X']).shape)
mean_displacement=mean_displacement_t(dat['X'])
plt.plot(mean_displacement, color='purple')
#plt.plot((mean_displacement)**2, color='orange')
```

The displacement at the n'th timestep of any individual particle is a sum of n (-1 and 1)-valued random variables X. The mean of the sum of these random variables is 0, because $0.5*(-1)+0.5*1=0$ for a single random variable and the mean of their sums is zero by Linearity of Expectation. However, this sum is a random variable and it fluctuates. The mean squared fluctuation of the sum is the variance of the sum. We can compute the variance of this sum: $Var(\sum_i{X_{i}})=\sum_i{Var(X)}=n$, because the variance of X is 1, because $(-1-0)^2*0.5+(1-0)^2*0.5)=1$ and each step is independent. The square root of variance of the sum is the standard deviation, which measures the average deviation of the sum of the random variables from the theoretical mean and it's $\sqrt{n}$ in this case. 

```python
#https://math.stackexchange.com/questions/1850653/mean-absolute-deviation-of-normal-distribution
#https://blog.arkieva.com/relationship-between-mad-standard-deviation/
plt.plot(mean_displacement, color='purple')
plt.plot(np.sqrt(2/np.pi)*np.sqrt(np.arange(0,1001)), color='blue')
plt.plot(np.mean(np.abs(dat['X'][:10,:]), axis=0), color='green')
#plt.plot(np.mean(np.abs(dat['X']), axis=0)/np.sqrt(np.arange(0,1001)), color='red')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
num_particles = 10000  # Number of particles
max_time = 1000  # End time of simulation in msec

# Initial position (column vector of 0's)
x0 = np.zeros(num_particles)

# Generate all displacements at once
dX = 2 * np.random.binomial(1, 0.5, (num_particles, max_time)) - 1

# Calculate cumulative sum to get positions
X = np.cumsum(dX, axis=1)
X = np.hstack((x0[:, np.newaxis], X)) 
```
