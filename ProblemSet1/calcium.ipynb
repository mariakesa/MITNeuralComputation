```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
R_tot=20*10**-9
L_start=0#microM
L_end=200*10**-6
K_D=np.array([10,30,50,70,90])*10**-6#microM
```

```python
L=np.arange(L_start,L_end, 0.1*10**-6)
for j in range(0,len(K_D)):
    K_d=K_D[j]*10**6
    LR=R_tot*L/(L+K_D[j])
    plt.plot(L,LR,label='K_D=%d' %K_d)
    plt.legend()
```
