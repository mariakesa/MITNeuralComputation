```python
from scipy.io import loadmat
import matplotlib.pyplot as plt
import numpy as np

dat=loadmat('MIT9_40S18_pset1/randomwalk.mat')
```

```python
print(dat['X'])
```

```python
plt.plot(dat['X'].T)
```

```python
print(dat['X'].shape)
```

```python
mean_by_t=np.mean(dat['X'], axis=0)
print(mean_by_t.shape)
plt.plot(mean_by_t)
```

```python
MSD=np.mean(dat['X']**2,axis=0)
plt.plot(MSD)
```

```python
#Fit slope

t=np.arange(0,1001)
print(t.shape)
slope=t@MSD/np.sum(t**2)
print(slope)
D=slope/2
print(D)
```
