```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
Rm = 100e6        # ohms
Cm = 100e-12      # farads
tau = Rm * Cm     # membrane time constant
dt = 1e-4
T = 0.5
t = np.arange(0, T, dt)

# Input: square wave current
I = np.zeros_like(t)
I[(t>0.1) & (t<0.2)] = 1e-9  # 1 nA pulse

# Membrane response (Euler integration)
V = np.zeros_like(t)
for i in range(1, len(t)):
    dV = dt/Cm * ( -V[i-1]/Rm + I[i-1] )
    V[i] = V[i-1] + dV

plt.plot(t, I*1e9, label="Injected current (nA)")
plt.plot(t, V*1e3, label="Membrane voltage (mV)")
plt.xlabel("Time (s)")
plt.legend()
plt.show()
```
